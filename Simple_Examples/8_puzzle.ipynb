```python
import math as Math
import numpy as np
import copy
# import pdb
```

```python

#As per the definition, the Manhattan the distance 
# is same as sum of the absolute difference of the coordinates.
#    Manhattan distance = |X1 – X2| + |Y1 – Y2|
class Puzz_Solver:
    def __init__(self,Init,Target):
            self.Init = Init
            self.Target = Target
            self.Final_States = []
            self.sizeA =  (int)(Math.sqrt(Init.size))
    ############################################################
    def conflict_calc(self,state):
            Num_of_conflicts = 0
            l = self.sizeA
            
            # Calculate the transpose of given Matrix:
            start_state_transpose = self.trans_calc(state)
            Final_state = np.transpose(self.Target)
            Final_state = Final_state.tolist()
            #shebahat satr ha va sotoon ha ra dar nazar migirad.
            for i in range(l):
                for j in range(l):
                    if self.check_conflicts(state[i], self.Target.tolist()[j]):
                        Num_of_conflicts = Num_of_conflicts +  1
                    if self.check_conflicts(start_state_transpose[i], Final_state[j]):
                        Num_of_conflicts = Num_of_conflicts+ 1

            return Num_of_conflicts
    ############################################################
    def trans_calc(self,a): # Transpose the list object
            l= self.sizeA
            temp = []
            transpose= []
            
            for i in range(l):
                for j in range(l):
                    temp.append(a[j][i])
                transpose.append(temp)
                temp = []
            return transpose
    ############################################################
    def check_conflicts(self,A,B):
            flag_0 = False
            flag_1 = False
            flag_2 = False
            l = len(A)
            for i in range(l):
                for j in range(l):
                    if A[i]==B[j]!=0:
                        if i-j<0:
                            flag_0 = True
                        if i-j>0:
                            flag_1 = True
                        if i==j==1:
                            flag_2 = True
            
            return (flag_2 and (flag_0 or flag_1)) or (flag_1 and flag_0)
            
    ############################################################    
    def h_calc(self,mode,Current):
        if(mode==1): # Just Manhattan Distance:
            Sum = 0
            for i in  range(self.sizeA):
                for j in range(self.sizeA):
                    X1 = i
                    Y1 = j
                    Places = np.argwhere(self.Target == Current[i][j])
                    X2  = Places[0][0]
                    Y2  = Places[0][1]
                    Sum = self.Manhattan_dist(X1,X2,Y1,Y2) + Sum
            return (int)(Sum/2)
        elif(mode==2):# Manhattan Distance + Linear Conflicts
            return self.h_calc(3,Current) + 2*self.conflict_calc(Current)
        elif(mode ==3):# Manhattan Distance in another way    
            Sum = 0
            for i in range(self.sizeA):
                for j in range(self.sizeA):
                    for l in range(self.sizeA):
                        for k in range(self.sizeA):
                            if(self.Target[i][j] == Current[l][k] != 0 ):
                                Sum = Sum + abs(i-l)+abs(j-k)
            return (int)(Sum)                    
        else:
            print("Mode Not Available!")
            return 0
    ############################################################
    def f_calc(self,g_val , h_val ):
        return   g_val +   h_val    
    ############################################################
    def check_next_Pos(self,row,col): # check mikonim k andise ma OK bashe va az matris biroon nazane
        bound = Math.sqrt(self.Init.size)
        if(row<0 or col<0 or row>bound-1 or col>bound-1): 
            return 0
        else:
            return 1  
    ############################################################        
    def Manhattan_dist(self,X1,X2,Y1,Y2):
        dist = abs(X2-X1)+abs(Y2-Y1)
        return (int)(dist)
    ############################################################
    def Check_end(self,Current , Goal ):
        if Current == Goal:
            return  1
        else:
            return 0
    ############################################################
    def Find_all_States(self, state , g_cost):
        
        my_list = state
        l = len(my_list)
        temp = []
        for i in range(l):
            for j in range(l):
                if my_list[i][j] == 0: # Next Possible Positions to change the place of 0
                    temp = [[i-1,j] , [i+1,j],
                            [i,j-1] , [i,j+1]]
                    index_i = i
                    index_j = j
                    break

        valid_positions = []
        t = my_list.copy()
        for position in temp:
            if self.check_next_Pos(position[0] , position[1]):
                valid_positions.append(position)
            Find_all_States = []

        for position in valid_positions:
            i, j = position[0], position[1]
            t = copy.deepcopy(my_list)
            t[index_i][index_j] = t[i][j]
            t[i][j] = 0
            Find_all_States.append(t)

        g_costs = [g_cost+1 for _ in Find_all_States]

        return Find_all_States, g_costs      
    #def Path_Finder(self):
    ############################################################
    def Draw_Matrix_beauty(self, Given_States ):
        print("                   ---------------")
        for Each_Step in Given_States:
            print("                    | "+str(Each_Step[0])+" | "+str(Each_Step[1])+" | "+str(Each_Step[2])+" | ")
        
    ############################################################    
    def draw_Out(self):
        n = self.Final_States
        counter = 0
        for Taken_Steps in n[::-1]:
            counter = counter +1
            print("Step Number: "+str(counter)  )
            self.Draw_Matrix_beauty(Taken_Steps)
            print("                   ###############")
            print("______________________________________________")
            print("                      ||                      ")
            print("                      ||                      ")
            print("                      \/                      ")
            
    ############################################################
    ############################################################
    ############################################################      
    
        
        
        
    def Solve(self): #  In Tabe Aslias ke seda zade mishe va miad baghie method haro call mikone ta hal kone!  
        print("Started The A* Algorithm")
        #pdb.set_trace()
        States = [ [self.Init.tolist()]  ] # Voroodio migire Assign mikone
        g_val = 0 # Noghte shoroo g ish 0 hast dg!
        f_val = [self.f_calc( g_val  , self.h_calc(2 , States[0][0] ) )]
        
        
        try:
            # Do Our Stuff
            time_order = 0
            spacial_order = 0
            g_costs = [0]
            Current  = States[0]
            #f_val = 
            while(1):# Goes on Until Convergence:
                
              
                
                
                n_states, n_g_costs = self.Find_all_States(Current[0] , g_val)
                
                a = Current.copy()
                del States[0]
                del g_costs[0]
                del f_val[0]
                 
                for state,g_cost in zip(n_states, n_g_costs):
                    b = a.copy()

                    if state not in a:
                        b.insert(0,state)
                        States.append(b)
                        time_order += 1
                        g_costs.append(g_cost)
                        f_val.append(self.f_calc(g_cost  , self.h_calc(2,state)  ))
                mapped = list(zip(f_val, g_costs, States))
                mapped.sort()  # Problem Here!
                t = list(zip(*mapped))  
                f_val = list(t[0])
                g_costs = list(t[1])  
                States = list(t[2])
                if(len(States)>spacial_order):
                    spacial_order = len(States)

                
                
                # Stopping Criteria:
                Current = States[0]
                g_val = g_costs[0]
                ENd = self.Check_end(Current[0] , self.Target.tolist())
                if (ENd == 1):
                    # The Result:
                    print("The Algorithm Converged!")
                    self.Final_States = States[0] 
                    #self.display_reult()
                    break
                
                
        except:
            print("There was something Wrong!")    
        
    



```
